### Initial Data Preprocessing
From foods.csv and beauty_products.csv extract all the food items which are tagged to be sold in germany. <br>
Also drop all the rows where a product name is missing. <br>
Only keep columns product name, countries and categories.

```python
import pandas as pd
```

### Foods

```python
### specify the columns
cols = [
    "product_name",
    "countries",
    "categories"
]

### load the csv file
df = pd.read_csv(
    "../data/foods.csv",
    sep="\t",
    usecols=cols,
    dtype=str,
    on_bad_lines="skip",
    low_memory=False
)
```

```python
### take only those products available in germany
df_german = df[df["countries"].str.contains("germany", case=False, na=False)]
### drop rows whre product name is NaN
df_german = df_german[df_german["product_name"].notna()]
```

```python
### save the csv
df_german.to_csv('../data/german_foods.csv', sep=',')
```

### Beauty Products

```python
### specify the columns
cols = [
    "product_name",
    "countries",
    "categories"
]

### load the csv file
df = pd.read_csv(
    "../data/beauty_products.csv",
    sep="\t",
    usecols=cols,
    dtype=str,
    on_bad_lines="skip",
    low_memory=False
)
```

```python
### take only those products available in germany
df_german = df[df["countries"].str.contains("germany", case=False, na=False)]
### drop rows whre product name is NaN
df_german = df_german[df_german["product_name"].notna()]
```

```python
df_german.to_csv('../data/german_beauty.csv', sep=',')
```
