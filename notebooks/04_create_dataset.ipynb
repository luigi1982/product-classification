### Combining the Foods and Beauty Datasets

```python
import pandas as pd
```

```python
### load both the datasets
df_foods = pd.read_csv("../data/german_foods.csv", sep=",")
df_beauty = pd.read_csv("../data/german_beauty.csv", sep=",")
```

```python
### drop redundant columns that where created during preprocessing
df_foods=df_foods.drop(columns=["Unnamed: 0"])
df_beauty=df_beauty.drop(columns=["Unnamed: 0", "Unnamed: 0.1"])
```

```python
### the csv holding the beauty products is missing the variable "edible"
### we set them all to False
df_beauty["edible"] = False
```

```python
### check if the columns match
assert list(df_beauty) == list(df_foods)
```

```python
### concatenate the two dataframes
df = pd.concat([df_foods, df_beauty])
```

```python
### save the dataframe
df.to_csv("../data/dataset.csv", sep=",")
```
